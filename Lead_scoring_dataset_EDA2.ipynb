## Treating Missing values

We will first look at the columns with relatively less missing values and probably drop the rows. We will then look at the columns with high number of missing values and by referring to the data dictionary or domain info from elsewhere we can either choose to drop the columns or impute the values accordingly

```python
# looking at 'Lead Source'

lead_df[lead_df["Lead Source"].isnull()]
```

```python
# Dropping these values as the percentage of missing values is less

lead_df = lead_df[~lead_df["Lead Source"].isnull()]
```

```python
lead_df.shape
```

```python
# looking at columns where the missing values percentage is less than 13%

"""
TotalVisits
Page Views Per Visit
Last Activity
"""

for col in ["TotalVisits", "Page Views Per Visit", "Last Activity"]:
    lead_df = lead_df[~lead_df[col].isnull()]

lead_df.shape
```

```python
# Looking at columns with missing values less than 30 %

"""
Country
Specialization                               
How did you hear about X Education           
What is your current occupation              
What matters most to you in choosing a course
"""
plt.figure(figsize=(8, 6))
sns.countplot(lead_df.Country)
plt.xticks(rotation=90)
plt.show()
```

# Model Building

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=100)
```

```python
X_train = df_train.drop('Converted', axis=1)
y_train = df_train['Converted']
```

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
```

```python
rfe = RFE(lr, n_features_to_select=18)
rfe.fit(X_train, y_train)
```

```python
rfe_df = pd.DataFrame(zip(X_train.columns, rfe.support_, rfe.ranking_), columns = ['Columns', 'Support', 'Rank'])
rfe_df[rfe_df.Support]
```

```python
columns = list(X_train.columns[rfe.support_])
```

```python
X_train_sm = X_train[columns]
X_train_sm = sm.add_constant(X_train_sm)

logit = sm.GLM(y_train, X_train_sm, family=sm.families.Binomial())
print(logit.fit().summary())
```

```python
from sklearn import metrics
```

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None

draw_roc(y_train, y_pred)
```

```python
y_pred_conv = np.where(y_pred > 0.3, 1, 0)
y_true = y_train
print('PRECISION')
print(precision_score(y_true, y_pred_conv))
print('RECALL')
print(recall_score(y_true, y_pred_conv))
```
